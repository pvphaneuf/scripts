```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
```

```python
df = pd.read_csv("./Untitled spreadsheet - Sheet1.csv", header=None)
df = df.fillna('')
df.head()
```

```python
list(df.iloc[1])
```

```python
o = ','.join(list(df.iloc[1])[2:4])
o = '"' + o + '"'
o
```

```python
def _reverse_strings_in_list(list_of_strings):
    rl = list()
    for s in list_of_strings:
        rl.append(''.join(reversed(s)))
    return rl


result = pd.DataFrame()
for _, reads in df.iterrows():
    
    # For the case where reads have the word "run" in them,
    # have to reverse the string for sorting generate the results I want.
    # After sorting, re-reverse to get the original strings.
    if "run" in reads[0]:
        sorted_reads = sorted(_reverse_strings_in_list(reads))
        sorted_reads = _reverse_strings_in_list(sorted_reads)
    else:
        sorted_reads = sorted(reads)
    
    # remove empty strings from list
    try:
        while True:
            sorted_reads.remove('')
    except ValueError:
        pass
    
    xpmd_entry = sorted_reads[:2]
    if len(sorted_reads) > 2:
        o = ','.join(sorted_reads[2:4])
        o = '"' + o + '"'
        xpmd_entry.append(o)
    
    result = result.append([xpmd_entry])


result.fillna('', inplace=True)
result.to_csv("./SvNS_organized_reads.csv")
result.head()
```
